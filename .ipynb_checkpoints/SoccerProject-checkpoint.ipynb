<b> Grabbing EPL clubs and their average Fifa Team Statistics </b>

```python
man_utd = team[team['team_short_name'] == 'MUN']
# man_utd team_api_id = 10260, team_fifa_api_id = 11
fifa = team.merge(team_attr,on='team_api_id')
fifa_mean = fifa.groupby(['team_long_name'])[['team_api_id','buildUpPlaySpeed','buildUpPlayDribbling','buildUpPlayPassing','chanceCreationPassing','chanceCreationCrossing','chanceCreationShooting','defencePressure','defenceAggression','defenceTeamWidth']].mean().round()
fifa_mean['team_api_id'] = fifa_mean['team_api_id'].astype(int)

fifa_mean_league = fifa_mean.merge(match,left_on='team_api_id',right_on='home_team_api_id')[['league_id','team_api_id','buildUpPlaySpeed','buildUpPlayDribbling','buildUpPlayPassing','chanceCreationPassing','chanceCreationCrossing','chanceCreationShooting','defencePressure','defenceAggression','defenceTeamWidth']]
fifa_mean_league = fifa_mean_league.groupby(['team_api_id']).mean()
fifa_mean_league = fifa_mean_league.merge(team,on='team_api_id')[['team_long_name','league_id','team_api_id','buildUpPlaySpeed','buildUpPlayDribbling','buildUpPlayPassing','chanceCreationPassing','chanceCreationCrossing','chanceCreationShooting','defencePressure','defenceAggression','defenceTeamWidth']]
fifa_mean_league.head()

epl_fifa_mean = fifa_mean_league[fifa_mean_league['league_id']==1729]
epl_fifa_mean.head()
```

<b> Grabbing EPL clubs and their total points accumulated </b>

```python
conditions = [
    (match['home_team_goal']>match['away_team_goal']),
    (match['home_team_goal']==match['away_team_goal']),
    (match['home_team_goal']<match['away_team_goal'])
]
values_home = [3,1,0]
values_away = [0,1,3]

match['points_home'] = np.select(conditions,values_home) 
match['points_away'] = np.select(conditions,values_away)

match_with_names_home = match.merge(team,how='inner',left_on='home_team_api_id', right_on='team_api_id')
match_with_names_away = match.merge(team,how='inner',left_on='away_team_api_id', right_on='team_api_id')

team_points_home = match_with_names_home.groupby(['team_long_name'])[['points_home']].sum()
team_points_away = match_with_names_away.groupby(['team_long_name'])[['points_away']].sum()

team_points = team_points_home.merge(team_points_away, on='team_long_name')
team_points['points_total'] = team_points['points_home'] + team_points['points_away']

team_id_points = team_points.merge(team,on='team_long_name')[['team_long_name','points_total','team_api_id']]
team_league_points = team_id_points.merge(match,left_on='team_api_id',right_on='home_team_api_id')[['team_long_name','points_total','team_api_id','league_id']]
team_league_points = team_league_points.groupby(['team_long_name'])[['points_total','league_id']].mean()

epl_points = team_league_points[team_league_points['league_id'] == 1729]
epl_points.head()
```

```python
epl_final = epl_points.merge(epl_fifa_mean,on='team_long_name')
epl_final = epl_final.drop(columns =['league_id_x','league_id_y','team_api_id'])
epl_final.head()

```

<b> Setting up for Feature Selection </b>

```python
import matplotlib
import matplotlib.pyplot as plt
import seaborn as sns
import statsmodels.api as sm
%matplotlib inline
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression, LogisticRegression
from sklearn.feature_selection import RFE
from sklearn.linear_model import RidgeCV, LassoCV, Ridge, Lasso
from sklearn.svm import SVC
from sklearn.neighbors import KNeighborsClassifier
from sklearn.discriminant_analysis import LinearDiscriminantAnalysis, QuadraticDiscriminantAnalysis
from sklearn.naive_bayes import GaussianNB
from sklearn import metrics
from sklearn.tree import DecisionTreeClassifier
from pandas.plotting import parallel_coordinates

fifa_mean_league.head()
team_league_points.head()
final = team_league_points.merge(fifa_mean_league,on='team_long_name')
final = final.drop(columns =['league_id_x','league_id_y','team_api_id','buildUpPlayDribbling'])
final = final.set_index('team_long_name')

X = final.drop('points_total',1)
y = final['points_total']
```

<b> Performing Pearson correlation </b>

```python
plt.figure(figsize=(12,10))
cor = final.corr()
sns.heatmap(cor, annot=True, cmap=plt.cm.Reds)
plt.show()
```

From Pearson correlation, we find that there are not really any strong correlations between the Fifa Team statistics and points total. However, the strongest correlation lies within buildUpPlayPassing and defencePressure. However, buildUpPlayPassing has a negative correlation so it should be lower, whereas defencePressure should be higher

```python
cor_target = abs(cor['points_total'])
relevant_features = cor_target[cor_target>.2]
relevant_features
```

<b>Performing Recursive Feature Elimination</b>

```python
model = LinearRegression()

rfe = RFE(model, 4)
X_rfe = rfe.fit_transform(X,y)

model.fit(X_rfe,y)
print(rfe.support_)
print(rfe.ranking_)    
```

<b> Find optimal number of features for RFE </b>

```python
nof_list = np.arange(1,8)
high_score=0
nof=0
score_list = []
for n in range(len(nof_list)):
    X_train, X_test, y_train, y_test = train_test_split(X,y,test_size=0.3,random_state=0)
    model = LinearRegression()
    rfe = RFE(model,nof_list[n])
    X_train_rfe = rfe.fit_transform(X_train,y_train)
    X_test_rfe = rfe.transform(X_test)
    model.fit(X_train_rfe,y_train)
    score = model.score(X_test_rfe,y_test)
    score_list.append(score)
    if (score>high_score):
        high_score = score
        nof = nof_list[n]

print(f"Optimum number of features: {nof}")
print(f'Score with {nof} features: {high_score}')
```

```python
cols = list(X.columns)
model = LinearRegression()

rfe = RFE(model, 6)

X_rfe = rfe.fit_transform(X,y)

model.fit(X_rfe,y)
temp = pd.Series(rfe.support_,index=cols)
selected_features = temp[temp==True].index
print(selected_features)
```

<b> Using Lasso Regression Model </b>

```python
reg = LassoCV()
reg.fit(X, y)
print(f'Best Alpha {reg.alpha_}')
print(f'Best Score {reg.score(X,y)}')
coef = pd.Series(reg.coef_,index=X.columns)
print('Lasso picked ' + str(sum(coef!=0)) + ' variables and eliminated ' + str(sum(coef == 0)) + ' variables')
```

```python
imp_coef = coef.sort_values()
matplotlib.rcParams['figure.figsize'] = (8.0,10.0)
imp_coef.plot(kind = 'barh')
plt.title('Feature importance using Lasso Model')
```

From these methods we find that the Fifa Team statistics do not have much correlation to total points accumulated. However, if you were forced to make a decision, the two strongest correlations are defencePressure (positively) and buildUpPlayPassing (negatively)

### Beginning of Odds Analysis

<b> Making Results Column </b>

```python
match_odds = match[['id','league_id','home_team_api_id','away_team_api_id','home_team_goal','away_team_goal','points_home','points_away','B365H','B365D','B365A','BWH','BWD','BWA','IWH','IWD','IWA','LBH','LBD','LBA','PSH','PSD','PSA','WHH','WHD','WHA','SJH','SJD','SJA','VCH','VCD','VCA','GBH','GBD','GBA','BSH','BSD','BSA']]
match_odds = match_odds.dropna()

def result(row):
    if row['home_team_goal'] == row['away_team_goal']:
        val = 'D'
    elif row['home_team_goal'] > row['away_team_goal']:
        val = 'H'
    else:
        val = 'A'
    return val

match_odds['result'] = match_odds.apply(result,axis=1)
match_odds.head()
```

```python
draw_odds = match_odds[match_odds['result'] == 'D']
home_win_odds = match_odds[match_odds['result'] == 'H']
away_win_odds = match_odds[match_odds['result']== 'A']

draw_odds = draw_odds[['B365D','BWD','IWD','LBD','PSD','WHD','SJD','VCD','GBD','BSD','result']]
home_win_odds = home_win_odds[['B365H','BWH','IWH','LBH','PSH','WHH','SJH','VCH','GBH','BSH','result']]
away_win_odds = away_win_odds[['B365A','BWA','IWA','LBA','PSA','WHA','SJA','VCA','GBA','BSA','result']]

draw_odds = draw_odds.groupby(['result']).sum()
home_win_odds = home_win_odds.groupby(['result']).sum()
away_win_odds = away_win_odds.groupby(['result']).sum()

draw_odds.insert(0, 'index', 0)
draw = draw_odds.set_index('index')
home_win_odds.insert(0,'index',0)
home = home_win_odds.set_index('index')
away_win_odds.insert(0,'index',0)
away = away_win_odds.set_index('index')

result = pd.DataFrame({'B365':[]})
result['B365'] = draw['B365D'] + home['B365H'] + away['B365A']
result['BW'] = draw['BWD'] + home['BWH'] + away['BWA']
result['IW'] = draw['IWD'] + home['IWH'] + away['IWA']
result['LB'] = draw['LBD'] + home['LBH'] + away['LBA']
result['PS'] = draw['PSD'] + home['PSH'] + away['PSA']
result['WH'] = draw['WHD'] + home['WHH'] + away['WHA']
result['SJ'] = draw['SJD'] + home['SJH'] + away['SJA']
result['VC'] = draw['VCD'] + home['VCH'] + away['VCA']
result['GB'] = draw['GBD'] + home['GBH'] + away['GBA']
result['BS'] = draw['BSD'] + home['BSH'] + away['BSA']
result
```

```python
row = result.iloc[0]
row.plot(kind='bar')
plt.ylim([7400,8100])
plt.show()
```

### Beginning of Player Analysis

<b> Showing/Finding Most Improved Players </b>

```python
player_merged = player.merge(player_attr, on='player_api_id')
player_merged = player_merged.drop(columns=['id_x','id_y','player_fifa_api_id_x','player_fifa_api_id_y'])
p = player_merged.groupby(['player_name']).agg({'date':[np.min,np.max]})

p_first = pd.merge(p, player_merged, left_on=['player_name',('date','amin')],right_on=['player_name','date'])[['player_name','date','overall_rating']]
p_last = pd.merge(p, player_merged, left_on=['player_name',('date','amax')],right_on=['player_name','date'])[['player_name','date','overall_rating']]
p_new = p_first.merge(p_last, on='player_name')
p_new['overall_increase'] = p_new['overall_rating_y'] - p_new['overall_rating_x']
p_new.head()
```

```python
p_new = p_new.dropna()
p_new.sort_values(by='overall_increase',axis=0, ascending=False)

plt.scatter(p_new['overall_increase'],p_new['overall_rating_y'])
plt.show()
```

<b> Classifying based on Players Positions </b>

```python
player_stats = player2.set_index('Name')
player_stats = player_stats.drop(columns=['ID','Age','Photo','Nationality','Flag','Club','Club Logo','Value','Wage','Special','Preferred Foot','International Reputation','Work Rate','Body Type','Real Face','Jersey Number','Joined','Loaned From','Contract Valid Until','Height','Weight','LS','ST','RS','LW','LF','CF','RF','RW','LAM','CAM','RAM','LM','LCM','CM','RCM','RM','LWB','LDM','CDM','RDM','RWB','LB','LCB','CB','RCB','RB','Release Clause'])

for index, row in player_stats.iterrows():
    ifor_val = 'GK'
    if row['Position'] == 'RF' or row['Position'] == 'LF' or row['Position'] == 'RS' or row['Position'] == 'LS' or row['Position'] == 'ST':
        ifor_val = 'ST'
    elif row['Position'] == 'RW' or row['Position'] == 'LW' or row['Position'] == 'RM' or row['Position'] == 'LM':
        ifor_val = 'W'
    elif row['Position'] == 'RCM' or row['Position'] == 'LCM' or row['Position'] == 'CM' or row['Position'] == 'RDM' or row['Position'] == 'LDM' or row['Position'] == 'CDM':
        ifor_val = 'CM'
    elif row['Position'] == 'RCB' or row['Position'] == 'LCB' or row['Position'] == 'CB':
        ifor_val = 'CB'
    elif row['Position'] == 'RWB' or row['Position'] == 'RB' or row['Position'] == 'LWB' or row['Position'] == 'LB':
        ifor_val = 'WB'
            
    
    player_stats.at[index,'Position'] = ifor_val

player_stats.head()
```

```python
player_stats = player_stats.dropna()
train, test = train_test_split(player_stats, test_size = 0.4, stratify = player_stats['Position'], random_state = 42)

X_train = train.drop(columns = ['Position'])
y_train = train.Position
X_test = test.drop(columns = ['Position'])
y_test = test.Position
```

```python
sns.violinplot(x="Position", y="StandingTackle", data=train, size=5, palette ='colorblind')
```

<b> Decision Tree Classifier </b>

```python
tree = DecisionTreeClassifier(random_state=1)
tree.fit(X_train,y_train)
prediction = tree.predict(X_test)

print(f'Accuracy = {metrics.accuracy_score(prediction,y_test)}')
print(tree.feature_importances_)
```

<b> Guassian Naive Bayes Classifier </b>

```python
gauss = GaussianNB()
y_pred = gauss.fit(X_train,y_train).predict(X_test)
print(f'Accuracy = {metrics.accuracy_score(y_pred,y_test)}')
```

<b> LDA Classifier </b>

```python
lda = LinearDiscriminantAnalysis()
y_pred = lda.fit(X_train,y_train).predict(X_test)
print(f'Accuracy = {metrics.accuracy_score(y_pred,y_test)}')
```

<b> QDA Classifier </b>

```python
qda = QuadraticDiscriminantAnalysis()
y_pred = qda.fit(X_train,y_train).predict(X_test)
print(f'Accuracy = {metrics.accuracy_score(y_pred,y_test)}')
```

<b> KNN Classifier </b>

```python
acc_s = pd.Series(dtype = 'float')
for i in list(range(1,11)):
    knn = KNeighborsClassifier(n_neighbors = i)
    knn.fit(X_train,y_train)
    pred = knn.predict(X_test)
    acc_s = acc_s.append(pd.Series(metrics.accuracy_score(pred,y_test)))
    
plt.plot(list(range(1,11)), acc_s)
plt.title('Test Accuracy vs K')
plt.xticks(list(range(1,11)))
```

<b> Logistic Regression </b>

```python
lr = LogisticRegression(solver='newton-cg').fit(X_train,y_train)
pred = lr.predict(X_test)
print(f'Accuracy = {metrics.accuracy_score(pred,y_test)}')
```
